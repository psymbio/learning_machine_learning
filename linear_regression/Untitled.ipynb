```python
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(figsize=(8, 5))

# Input nodes
input_positions = np.array([(0.5, 4), (0.5, 3), (0.5, 1)])

for pos in input_positions:
    circle = plt.Circle(pos, 0.2, color='w', ec='black', zorder=2)
    ax.add_patch(circle)

# Weight nodes
weight_positions = np.array([(2, 4), (2, 3), (2, 1)])

for pos in weight_positions:
    circle = plt.Circle(pos, 0.2, color='w', ec='black', zorder=2)
    ax.add_patch(circle)

# Lines connecting inputs to weights
for i in range(len(input_positions)):
    ax.plot([input_positions[i][0], weight_positions[i][0]], [input_positions[i][1], weight_positions[i][1]], 'k-')

# Sum node
sum_pos = (3.5, 2.5)
sum_circle = plt.Circle(sum_pos, 0.3, color='w', ec='black', zorder=2)
ax.add_patch(sum_circle)
ax.text(3.5, 2.5, 'Σ', fontsize=14, ha='center', va='center')

# Lines connecting weights to sum
for pos in weight_positions:
    ax.plot([pos[0], sum_pos[0]], [pos[1], sum_pos[1]], 'k-')

# Bias node
bias_pos = (3.5, 4)
bias_circle = plt.Circle(bias_pos, 0.2, color='w', ec='black', zorder=2)
ax.add_patch(bias_circle)
ax.text(3.5, 4, 'b', fontsize=12, ha='center', va='center')
ax.plot([bias_pos[0], sum_pos[0]], [bias_pos[1], sum_pos[1]], 'k-')

# Activation function node
act_pos = (5, 2.5)
rect = plt.Rectangle((4.6, 2.2), 0.8, 0.6, color='w', ec='black')
ax.add_patch(rect)
ax.text(5, 2.5, 'φ(·)', fontsize=14, ha='center', va='center')

# Line connecting sum to activation function
ax.plot([sum_pos[0], act_pos[0]], [sum_pos[1], act_pos[1]], 'k-')

# Output node
output_pos = (6.5, 2.5)
ax.plot([act_pos[0], output_pos[0]], [act_pos[1], output_pos[1]], 'k-')
ax.text(6.5, 2.5, 'y', fontsize=12, ha='center', va='center')

# Labels
ax.text(0.1, 4, '$x_1$', fontsize=12)
ax.text(0.1, 3, '$x_2$', fontsize=12)
ax.text(0.1, 1, '$x_m$', fontsize=12)
ax.text(2, 4.3, '$ω_1$', fontsize=12)
ax.text(2, 3.3, '$ω_2$', fontsize=12)
ax.text(2, 1.3, '$ω_m$', fontsize=12)

# Axis settings
ax.axis('off')
ax.set_xlim(0, 7)
ax.set_ylim(0, 5)
plt.show()
```
